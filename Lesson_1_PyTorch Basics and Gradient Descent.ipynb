# Linear Regression with PYTORCH

```python
import numpy as np
import torch
```

```python
inputs=np.array([
                 [73,67,43],
                 [91,88,64],
                 [87,134,58],
                 [102,43,37],
                 [69,96,70]
],dtype="float32")
```

```python
targets=np.array([
                  [56,70],
                  [81,101],
                  [199,113],
                  [22,37],
                  [103,119]
                  ],dtype="float32")
```

```python
inputs=torch.from_numpy(inputs)
targets=torch.from_numpy(targets)
```

```python
w=torch.randn(2,3,requires_grad=True)
b=torch.randn(2,requires_grad=True)
print(w,b)
```

```python
def model(x):
  return x @ w.t() + b
```

```python
preds=model(inputs)
print(preds)
```

```python
print(targets)
```

# Loss functions

```python
diff=targets-preds
diff
```

```python
def mse ( t1,t2):
  diff=t1-t2
  return torch.sum(diff*diff)/diff.numel()
```

```python
loss=mse(preds,targets)
loss
```

```python
#compute grads
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
print(b)
print(b.grad)
```

```python
with torch.no_grad():
  w-=w.grad*1e-5
  b-=b.grad*1e-5
  w.grad.zero_()
  b.grad.zero_()
```

```python
preds=model(inputs)
print(preds)
loss=mse(preds,targets)
loss
```

# Multiple epochs

```python
for i in range(1,100):
  preds=model(inputs)
  loss=mse(preds,targets)
  print('epch no :',i,', loss :',loss)
  loss.backward()
  with torch.no_grad():

    w-=w.grad*1e-4
    b-=b.grad*1e-4
    w.grad.zero_()
    b.grad.zero_()


```

# LR with Pytorch built ins 

```python
import torch.nn as nn
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70], 
                   [74, 66, 43], 
                   [91, 87, 65], 
                   [88, 134, 59], 
                   [101, 44, 37], 
                   [68, 96, 71], 
                   [73, 66, 44], 
                   [92, 87, 64], 
                   [87, 135, 57], 
                   [103, 43, 36], 
                   [68, 97, 70]], 
                  dtype='float32')

# Targets (apples, oranges)
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119],
                    [57, 69], 
                    [80, 102], 
                    [118, 132], 
                    [21, 38], 
                    [104, 118], 
                    [57, 69], 
                    [82, 100], 
                    [118, 134], 
                    [20, 38], 
                    [102, 120]], 
                   dtype='float32')
```

```python
inputs=torch.from_numpy(inputs)
targets=torch.from_numpy(targets)
```

```python
from torch.utils.data import TensorDataset
```

```python
train_ds=TensorDataset(inputs,targets)
train_ds[0:3]
```

```python
from torch.utils.data import DataLoader
```

```python
batch_size=5
train_dl=DataLoader(train_ds,batch_size,shuffle=True)
```

```python
for xb,yb in train_dl:
  print(xb)
  print(yb)
  print('batch')
```

```python
model=nn.Linear(3,2)
print(model.weight)
print(model.bias)
```

```python
list(model.parameters())
```

```python
preds=model(inputs)
preds
```

```python
import torch.nn.functional as F
```

```python
loss_fn=F.mse_loss
```

```python
loss=loss_fn(model(inputs),targets)
loss
```

```python
opt=torch.optim.SGD(model.parameters(),lr=1e-5)
```

```python
def fit (num_epochs,model,loss_fn,opt):
  for epoch in range (0,num_epochs):
    for xb,yb in train_dl:
      preds=model(xb)
      loss=loss_fn(preds,yb)
      loss.backward()
      opt.step()
      opt.zero_grad()
    if(epoch+1)%10==0:
      print('Epoch [{}/{}], Loss : {:.4f}'.format(epoch+1,num_epochs,loss.item()))
```

```python
fit(100,model,loss_fn,opt)
```
